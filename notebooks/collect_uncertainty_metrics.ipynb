```python
import uncertainty_toolbox as uct
import pandas as pd
from pathlib import Path
```

```python
out_dir = Path("./experiments")
```

```python
accuracy_results = []
for filepath in out_dir.rglob("testset_predictions.csv"):
    [_, lds, dataset, split, fold, ue, _] = str(filepath).split("/")
    df = pd.read_csv(filepath)
    if ue == "JQR":
        y_pred_col = "Y_POSTERIOR_MEAN"
    else:
        y_pred_col = "Y_PRED_MEAN"
    try:
        accuracy_metrics = uct.get_all_accuracy_metrics(y_pred=df["Y_PRED_MEAN"].values, y_true=df["Y_TRUE"].values, verbose=False)
        accuracy_metrics["LDS"] = lds.split("-")[-1]
        accuracy_metrics["DATASET"] = dataset
        accuracy_metrics["SPLIT"] = split
        accuracy_metrics["fold"] = int(fold[4:])
        accuracy_metrics["ue"] = ue
        accuracy_results.append(accuracy_metrics)
    except KeyError:
        print(filepath)
```
